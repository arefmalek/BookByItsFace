```python
import torch
import torchvision
import os
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision.transforms as transforms
from torchvision.transforms import ToTensor
%matplotlib inline

import pycuda.driver as cuda
cuda.init()
## Get Id of default device
torch.cuda.current_device()
# 0
```

```python
device = cuda.Device(0)
```

```python
class ImageFolderWithPaths(torchvision.datasets.ImageFolder):
    """Custom dataset that includes image file paths. Extends
    torchvision.datasets.ImageFolder
    """

    # override the __getitem__ method. this is the method that dataloader calls
    def __getitem__(self, index):
        # this is what ImageFolder normally returns 
        original_tuple = super(ImageFolderWithPaths, self).__getitem__(index)
        # the image file path
        path = self.imgs[index][0]
        # make a new tuple that includes original and the path
        tuple_with_path = (original_tuple + (path,))
        return tuple_with_path
```

```python
# We can use an image folder dataset the way we have it setup.
# Create the dataset
dataset = ImageFolderWithPaths(root="data/", transform=ToTensor())

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
trainset, testset = torch.utils.data.random_split(dataset, [train_size, test_size])

# Create the dataloader
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4,
                                         shuffle=True, num_workers=2)
testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                         shuffle=True, num_workers=2)
```

```python
def find_race(filename):
    demographics = filename.split("_")
    race = int(demographics[2] )
    return race
def find_age(filename):
    demographics = filename.split("_")[0]
    age = int(demographics[demographics.rfind('/')+1:])
    return age
def find_gender(filename):
    demographics = filename.split("_")
    gender = int(demographics[1] )
    return gender
```

```python
import matplotlib.pyplot as plt
import numpy as np

# functions to show an image


def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(testloader)
images, idk, filename = dataiter.next()
# show images
imshow(torchvision.utils.make_grid(images))

for fname in filename:
    print(find_race(fname), find_age(fname))
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.cuda()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 94 * 94, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84,4)  #age (0 to 1 * 100), gender (<0.5 -> F, >=0.5 -> M), race ((0 to 1) / nraces) 

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 94 * 94)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
age = Net()
race = Net()
gender = Net()

age.to(torch.device("cuda:0"))
race.to(torch.device("cuda:0"))
```

```python
criterion = nn.MSELoss()
race_optimizer = optim.Adam(race.parameters())
age_optimizer = optim.Adam(age.parameters())
```

```python
def generate_map(dataloader, func):
    demographic_map = {}
    for (i, (_, _, labels)) in enumerate(dataloader):
        for label in labels:
            try:
                demographic_map[label] = func(label)
            except ValueError:
                #this only happens on race map
                print("bad data: {}".format(label))
                demographic_map[label] = 4
    return demographic_map
```

```python
train_age = generate_map(trainloader, find_age)
```

```python
train_race = generate_map(trainloader, find_race)
```

```python
#Training the network for race :O
for epoch in range(10):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, _, labels = data

        #hashmaps become real neato here 
        races = [train_race[label] for label in labels]        
        
        # zero the parameter gradients
        race_optimizer.zero_grad()
        
        #convert list to tensor for PT functions
        labels = torch.Tensor(races)
        
        try:
            labels = torch.reshape(labels, (4,))
        except RuntimeError:
            print(labels, races)

        # forward + backward + optimize
        outputs = race(inputs.cuda())
        outputs = torch.reshape(outputs, (4, ))

        loss = criterion(outputs, labels.cuda())
        loss.backward()
        race_optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 500 == 0:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.8f' %
                  (epoch + 1, i + 1, running_loss / 500))
            running_loss = 0.0

print('Finished Training')
```

```python
#Training the network for age :O
for epoch in range(10):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, _, labels = data

        #hashmaps become real neato here 
        ages = [train_age[label] for label in labels]        
        
        # zero the parameter gradients
        age_optimizer.zero_grad()
        
        #convert list to tensor for PT functions
        labels = torch.Tensor(ages)
        
        labels = torch.reshape(labels, (4,))

        # forward + backward + optimize
        outputs = age(inputs.cuda())
        outputs = torch.reshape(outputs, (4, ))

        loss = criterion(outputs, labels.cuda())
        loss.backward()
        age_optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 500 == 0:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.8f' %
                  (epoch + 1, i + 1, running_loss / 500))
            running_loss = 0.0

print('Finished Training')
```
